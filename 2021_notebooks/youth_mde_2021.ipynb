```python
import pandas as pd
# To install, pip install tabula-py in your env
import tabula
```

```python
tabula.convert_into(r"..\resources\2021_data.pdf", r"..\uncleaned_csv\youth_mde_2021.csv", output_format="csv", pages='23', stream = True)
```

```python
# Read in csv
youth_mde_df = pd.read_csv(r"..\uncleaned_csv\youth_mde_2021.csv")

youth_mde_df = youth_mde_df.dropna(how='all')
```

```python
youth_mde_df
```

```python
# Extract second half of ami_df to make second df to be joined back in
youth_mde_df_2 = youth_mde_df[["Rank", "State", "%.1", "#.1"]]
youth_mde_df_2 = youth_mde_df_2.rename(columns={"%.1" : "%", "#.1": "#"})
```

```python
# drop seperated columns
youth_mde_df = youth_mde_df.drop(columns={"Rank", "State", "%.1", "#.1"})
```

```python
# Splitup problematic data
youth_mde_df['Rank'], youth_mde_df['State'] = youth_mde_df['Rank State'].str.split(' ', 1).str

youth_mde_df = youth_mde_df.drop(columns='Rank State')
```

```python
# Rename and reappend cleaned half of dataframe.
youth_mde_df = youth_mde_df[['Rank', 'State', '%', '#']]
youth_mde_df = youth_mde_df.append(youth_mde_df_2)
```

```python
# Reset index to make cleaning possible
youth_mde_df = youth_mde_df.reset_index(drop=True)
```

```python
youth_mde_df
```

```python
# Assign proper values
youth_mde_df.loc[0, "Rank"]=1
youth_mde_df.loc[0, "State"]='Mississippi'
youth_mde_df.loc[26, "#"] = 41000
```

```python
# Add suffix 
youth_mde_df = youth_mde_df.add_suffix('_youth_mde_2021')
```

```python
youth_mde_df = youth_mde_df.rename(columns={'State_youth_mde_2021' : 'State'})
```

```python
youth_mde_df.to_csv(r'..\resources\youth_mde_clean2021.csv')
```

```python
import pandas as pd
# To install, pip install tabula-py in your env
import tabula
```

```python
tabula.convert_into(r"resources\2020_data.pdf", "uncleaned_csv\youth_ami_2020_unclean.csv", output_format="csv", pages='31')
```

```python
# Read in csv
youth_ami_df = pd.read_csv("uncleaned_csv\youth_ami_2020_unclean.csv")
youth_ami_df.head()
```

```python
# Extract second half of ami_df to make second df to be joined back in
youth_ami_2_df = youth_ami_df[["Rank", "State", "%.1", "#.1"]]
youth_ami_2_df = youth_ami_2_df.rename(columns={ "%.1" : "%", "#.1": "#"})
youth_ami_2_df
```

```python
youth_ami_df = youth_ami_df.drop(columns={"Rank", "State", "%.1", "#.1"})
```

```python
youth_ami_df.head()
```

```python
youth_ami_df['Rank'], youth_ami_df['State'] = youth_ami_df['Rank State'].str.split(' ', 1).str

youth_ami_df.drop(columns='Rank State')
```

```python
youth_ami_df = youth_ami_df[['Rank', 'State', '%', '#']]
```

```python
youth_ami_df = youth_ami_df.append(youth_ami_2_df)
```

```python
youth_ami_df = youth_ami_df.dropna(how='all')
```

```python
youth_ami_df.loc[17, "Rank"]=16
youth_ami_df.loc[17, "State"]='South Dakota'
youth_ami_df.loc[17, "#"]=8000
```

```python
youth_ami_df = youth_ami_df.add_suffix('_youth_ami_2020')
```

```python
youth_ami_df.to_csv(r'resources\youth_ami_csv_clean_2020.csv', sep=',', index=False)
```

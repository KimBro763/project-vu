```python
import pandas as pd
# To install, pip install tabula-py in your env
import tabula
```

```python
tabula.convert_into(r"..\resources\2021_data.pdf", r"..\uncleaned_csv\youth_no_mhunclean2021.csv", output_format="csv", pages='28', stream = True)
```

```python
youth_nomh_df = pd.read_csv(r"..\uncleaned_csv\youth_no_mhunclean2021.csv")
youth_nomh_df = youth_nomh_df.dropna(how='all')
```

```python
# reset Index
youth_nomh_df.reset_index(drop=True)
```

```python
# Drop bugged rows
youth_nomh_df = youth_nomh_df.drop(youth_nomh_df.index[[26,27,28,29,30]])
```

```python
# Extract second half of ami_df to make second df to be joined back in
youth_nomh_df_2 = youth_nomh_df[["Rank.1", "State.1", "%.1", "#.1"]]
youth_nomh_df_2 = youth_nomh_df_2.rename(columns={'Rank.1' : "Rank", "State.1" : "State", "%.1" : "%", "#.1": "#"})
```

```python
youth_nomh_df = youth_nomh_df.drop(columns={"Rank.1", "State.1", "%.1", "#.1"})
```

```python
# Merge both dfs together, making final df
youth_nomh_df = youth_nomh_df.append(youth_nomh_df_2)
```

```python
# change dtype of # column
youth_nomh_df['#'] = youth_nomh_df['#'].str.replace(',', '').astype(float)
```

```python
youth_nomh_df = youth_nomh_df.add_suffix('_youth_no_mh_care2021')
```

```python
youth_nomh_df = youth_nomh_df.rename(columns={'State_youth_no_mh_care2021' : 'State'})
```

```python
youth_nomh_df.to_csv(r'..\resources\youth_no_mh_service2021.csv', sep=',', index=False)
```

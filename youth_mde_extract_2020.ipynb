```python
import pandas as pd
# To install, pip install tabula-py in your env
import tabula
```

```python
tabula.convert_into(r"resources\2020_data.pdf", "uncleaned_csv\youth_mde_2020_unclean.csv", output_format="csv", pages='33')
```

```python
# Read in csv
youth_mde_df = pd.read_csv("uncleaned_csv\youth_mde_2020_unclean.csv")
youth_mde_df.dropna(how='all')
```

```python
# Extract second half of ami_df to make second df to be joined back in
youth_mde_df_2 = youth_mde_df[["Rank.1", "State.1", "%.1", "#.1"]]
youth_mde_df_2 = youth_mde_df_2.rename(columns={'Rank.1' : "Rank", "State.1" : "State", "%.1" : "%", "#.1": "#"})
```

```python
# Drop columns that have been extracted
youth_mde_df = youth_mde_df.drop(columns={"Rank.1", "State.1", "%.1", "#.1"})
```

```python
youth_mde_df = youth_mde_df.append(youth_mde_df_2)
```

```python
youth_mde_df = youth_mde_df.dropna(how='all')
```

```python
youth_mde_df = youth_mde_df.add_suffix('_youth_mde_2020')
```

```python
youth_mde_df = youth_mde_df.rename(columns={'State_youth_mde_2020' : 'State'})
```

```python
youth_mde_df.to_csv(r'resources\youth_mde_clean2020.csv', sep=',', index=False)
```

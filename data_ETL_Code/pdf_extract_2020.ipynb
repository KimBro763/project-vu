```python
import pandas as pd
# To install, pip install tabula-py in your env
import tabula
```

```python
tabula.convert_into(r"resources\2020_data.pdf", "uncleaned_csv\output_2020.csv", output_format="csv", pages='28', stream = True)
```

```python
# Read in csv
ami_df = pd.read_csv("uncleaned_csv\output_2020.csv")
ami_df.head()
```

```python
# Extract second half of ami_df to make second df to be joined back in
ami_2_df = ami_df[["Rank.1", "State.1", "%.1", "#.1"]]
ami_2_df = ami_2_df.rename(columns={'Rank.1' : "Rank", "State.1" : "State", "%.1" : "%", "#.1": "#"})
ami_2_df
```

```python
# Drop columns that have been extracted
ami_df = ami_df.drop(columns={"Rank.1", "State.1", "%.1", "#.1"})
```

```python
# Merge both dfs together, making final df
ami_df = ami_df.append(ami_2_df)
```

```python
# reset index
ami_df = ami_df.reset_index(drop=True)
```

```python
# Drop NaN rows
ami_df = ami_df.drop(ami_df.index[[19,46]])
```

```python
# Reset index again
ami_df = ami_df.reset_index(drop=True)
```

```python
ami_df = ami_df.add_suffix('_adult_ami_2020')
```

```python
ami_df = ami_df.rename(columns={'State_adult_ami_2020' : 'State'})
```

```python
ami_df.to_csv(r'resources\ami_csv2020.csv', sep=',', index=False)
```

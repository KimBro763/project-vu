```python
import pandas as pd
from unicodedata import normalize
```

```python
# Select first dataframe from below source
df = pd.read_html('https://www.cdc.gov/mmwr/volumes/69/wr/mm6932a1.htm#T1_down')[0]
df.head()
```

```python
df = df.MultiIndex.drop_duplicates(keep='last')
```

```python
# function to clean up unnecessary whitespace retrieved from https://pbpython.com/pandas-html-table.html
def clean_normalize_whitespace(x):
    if isinstance(x, str):
        return normalize('NFKC', x).strip()
    else:
        return x
```

```python
# Runs clean_normalize_whitespace on main dataframe
new_df = df.applymap(clean_normalize_whitespace)
```

```python
# Create gender dataframe
gender_df = new_df[2:5]
gender_df
```

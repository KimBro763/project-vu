```python
import pandas as pd
# To install, pip install tabula-py in your env
import tabula
```

```python
tabula.convert_into(r"..\resources\2021_data.pdf", r"..\uncleaned_csv\adult_uninsured2021.csv", output_format="csv", pages='26', stream = True)
```

```python
# Read in csv
adult_uninsured_df = pd.read_csv(r"..\uncleaned_csv\adult_uninsured2021.csv")

adult_uninsured_df = adult_uninsured_df.dropna(how='all')
```

```python
adult_uninsured_df = adult_uninsured_df[pd.to_numeric(adult_uninsured_df['Adults with AMI who are Uninsured Rank'], errors='coerce').notnull()]
```

```python
adult_uninsured_df = adult_uninsured_df.rename(columns={'Adults with AMI who are Uninsured Rank' : 'Rank'})
```

```python
adult_uninsured_df
```

```python
adult_uninsured_df.loc[len(adult_uninsured_df.index)] = [None, 'National', 10.8, 5114000] 
```

```python
# change dtype of # column
adult_uninsured_df['#'] = adult_uninsured_df['#'].str.replace(',', '').astype(float)
```

```python
# Add suffix 
adult_uninsured_df = adult_uninsured_df.add_suffix('_adult_uninsured_2021')
```

```python
adult_uninsured_df = adult_uninsured_df.rename(columns={'State_adult_uninsured_2021' : 'State'})
```

```python
adult_uninsured_df.to_csv(r'..\resources\adult_uninsured_clean2021.csv')
```

```python
import pandas as pd
# To install, pip install tabula-py in your env
import tabula
```

```python
tabula.convert_into(r"..\resources\2021_data.pdf", r"..\uncleaned_csv\youth_some_mh_unclean2021.csv", output_format="csv", pages='29', stream = True)
```

```python
youth_someMH_df = pd.read_csv(r"..\uncleaned_csv\youth_some_mh_unclean2021.csv")
youth_someMH_df = youth_someMH_df.dropna(how='all')
```

```python
youth_someMH_df
```

```python
# Extract second half of ami_df to make second df to be joined back in
youth_someMH_df_2 = youth_someMH_df[["Rank", "State", "%.1", "#.1"]]
youth_someMH_df_2 = youth_someMH_df_2.rename(columns={"%.1" : "%", "#.1": "#"})
```

```python
# drop seperated columns
youth_someMH_df = youth_someMH_df.drop(columns={"Rank", "State", "%.1", "#.1"})
```

```python
# Splitup problematic data
youth_someMH_df['Rank'], youth_someMH_df['State'] = youth_someMH_df['Rank State'].str.split(' ', 1).str

youth_someMH_df = youth_someMH_df.drop(columns='Rank State')
```

```python
# Rename and reappend cleaned half of dataframe.
youth_someMH_df = youth_someMH_df[['Rank', 'State', '%', '#']]
youth_someMH_df = youth_someMH_df.append(youth_someMH_df_2)
```

```python
youth_someMH_df = youth_someMH_df.reset_index(drop=True)
```

```python
# Assign proper values
youth_someMH_df.loc[21, "Rank"]=22
youth_someMH_df.loc[21, "State"]='Arkansas'
youth_someMH_df.loc[47, "#"] = 15000
```

```python
# change dtype of # column
youth_someMH_df['#'] = youth_someMH_df['#'].str.replace(',', '').astype(float)
```

```python
youth_someMH_df = youth_someMH_df.add_suffix('_youth_someMH_2021')
```

```python
youth_someMH_df = youth_someMH_df.rename(columns={'State_youth_someMH_2021' : 'State'})
```

```python
youth_someMH_df.to_csv(r'..\resources\youth_some_mh_service_clean2021.csv', sep=',', index=False)
```

```python
import pandas as pd
# To install, pip install tabula-py in your env
import tabula
```

```python
tabula.convert_into(r"..\resources\2021_data.pdf", r"..\uncleaned_csv\adult_unmet_needs_2021.csv", output_format="csv", pages='25', stream = True)
```

```python
# Read in csv
adult_unmet_need = pd.read_csv(r"..\uncleaned_csv\adult_unmet_needs_2021.csv")

adult_unmet_need = adult_unmet_need.dropna(how='all')
```

```python
# Extract second half of ami_df to make second df to be joined back in
adult_unmet_need_2 = adult_unmet_need[["Rank.1", "State.1", "%.1", "#.1"]]
adult_unmet_need_2 = adult_unmet_need_2.rename(columns={'Rank.1' : "Rank", "State.1" : "State", "%.1" : "%", "#.1": "#"})
```

```python
# Drop columns that have been extracted
adult_unmet_need = adult_unmet_need.drop(columns={"Rank.1", "State.1", "%.1", "#.1"})
```

```python
adult_unmet_need
```

```python
# Rename and reappend cleaned half of dataframe.
adult_unmet_need = adult_unmet_need.append(adult_unmet_need_2)
```

```python
# Reset index to make cleaning possible
adult_unmet_need = adult_unmet_need.reset_index(drop=True)
```

```python
# Add suffix 
adult_unmet_need = adult_unmet_need.add_suffix('_adult_unmet_need_2021')
```

```python
adult_unmet_need = adult_unmet_need.rename(columns={'_adult_unmet_need_2021' : 'State'})
```

```python
adult_unmet_need.to_csv(r'..\resources\adult_unmet_need_clean2021.csv')
```
